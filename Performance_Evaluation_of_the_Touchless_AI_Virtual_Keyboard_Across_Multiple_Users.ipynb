```python
import matplotlib.pyplot as plt
import numpy as np

# -----our data -----
users      = ["User 1", "User 2", "User 3"]
accuracy   = [100, 95, 100]     # %
fps        = [22, 25, 21]       # frames per second
err_no_db  = [15, 12, 10]       # % error without debounce
err_db     = [5, 4, 3]          # % error with debounce

# ----- Plot -----
fig, axes = plt.subplots(3, 1, figsize=(7, 10), constrained_layout=True)

# 1) Accuracy
axes[0].bar(users, accuracy, color='skyblue')
axes[0].set_title("Gesture Recognition Accuracy per User")
axes[0].set_ylabel("Accuracy (%)")
axes[0].set_ylim(0, 105)
for x, y in zip(users, accuracy):
    axes[0].text(x, y + 1, f"{y}%", ha='center', va='bottom')

# 2) FPS
axes[1].bar(users, fps, color='lightgreen')
axes[1].set_title("Frame Rate (FPS) per User")
axes[1].set_ylabel("FPS")
axes[1].set_ylim(0, max(fps) + 5)
for x, y in zip(users, fps):
    axes[1].text(x, y + 0.5, f"{y}", ha='center', va='bottom')

# 3) Error rate comparison (grouped bars)
x = np.arange(len(users))
width = 0.35
axes[2].bar(x - width/2, err_no_db, width, label='Without Debounce', color='salmon')
axes[2].bar(x + width/2, err_db,    width, label='With Debounce',    color='orange')
axes[2].set_title("Error Rate Comparison")
axes[2].set_ylabel("Error Rate (%)")
axes[2].set_xticks(x)
axes[2].set_xticklabels(users)
axes[2].legend()
axes[2].set_ylim(0, max(err_no_db) + 5)
for xv, y1, y2 in zip(x, err_no_db, err_db):
    axes[2].text(xv - width/2, y1 + 0.3, f"{y1}%", ha='center', va='bottom', fontsize=9)
    axes[2].text(xv + width/2, y2 + 0.3, f"{y2}%", ha='center', va='bottom', fontsize=9)

plt.show()
# Optionally save:
# fig.savefig("gesture_metrics.png", dpi=300, bbox_inches="tight")
```
